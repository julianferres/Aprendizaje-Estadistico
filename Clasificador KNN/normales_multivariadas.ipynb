## Enunciado

Simular $\frac{n}{2}$ puntos de las normales multivariadas:

$ X_1 \sim \mathcal{N_2}( \mu_1,\,\Sigma ) $ 

$ X_2 \sim \mathcal{N_2}( \mu_2,\,\Sigma ) $ 


siendo:
- $n = 10, 100, 1000, 10000, \ldots$
- $\mu_1 = (-1,0)$ 
- $\mu_2 = (1,0)$
- $\Sigma = \mathbb{1}$ (matriz identidad)

## Soluci칩n

```python
#Import libraries
import numpy as np
import pandas as pd
#Plots
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
n = 1000 #Tama침o de muestra
```

```python
muestra = np.zeros((n,3))
```

#### Toma de muestra

```python
mean1 = [-1, 0]
mean2 = [1, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance

x1, y1 = np.random.multivariate_normal(mean1, cov, int(n/2)).T
x2, y2 = np.random.multivariate_normal(mean2, cov, int(n/2)).T
```

```python
x1 = pd.Series(x1, name="$X_1$")
x2 = pd.Series(x2, name="$X_2$")
y1 = pd.Series(y1, name="$Y_1$")
y2 = pd.Series(y2, name="$Y_2$")
```

```python
f, ax = plt.subplots(ncols = 2 , figsize=(16, 8))

g = ax[0].scatter( x1, y1, alpha='0.5', color='darkblue' ,  label = 'Clase 0');
g = ax[0].scatter( x2, y2, alpha='0.5', color='red' , label = 'Clase 1');
ax[0].legend()

# Draw the two density plots
ax[1] = sns.kdeplot(x1, y1,
                 cmap= "Blues", shade=True, shade_lowest=False)
ax[1] = sns.kdeplot(x2, y2,
                 cmap="Reds", shade=True, shade_lowest=False)
```

```python
g = sns.jointplot(x1, y1, color = 'lightblue' , kind="kde", space=0)
```

```python
g = sns.jointplot(x2, y2, color = 'red' , kind="kde", space=0)
```
