```python
#Import libraries
import numpy as np
#Plots
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
n = 1000 #Tama침o de muestra
```

#### Simulaci칩n de la distribucion sobre el triangulo con base en el eje x

```python
y1 = np.random.uniform( 0, 1, n )

x1 = np.zeros((n,))

for i in range(n):
    x1[i] = np.random.uniform((1-y1[i])/2 ,(1+y1[i])/2)
    
```

#### Simulaci칩n de la distribuci칩n sobre el triangulo con base en y=1

```python
y2 = np.random.uniform( 0, 1, n )

x2 = np.zeros((n,))

for i in range(n):
    x2[i] = np.random.uniform( y2[i]/2 , 1 - y2[i]/2)
```

```python
g = plt.scatter( x1 , y1 , alpha='0.5', color='lightgreen', label = 'Clase 0');
g = plt.scatter( x2 , y2 , alpha='0.5', color='orangered', label = 'Clase 1');
plt.legend();
plt.title("Distribuciones", fontsize=20)
```
