```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

#### Me genero 2 clusters para clasificar

```python
from sklearn.datasets.samples_generator import make_blobs 
```

```python
# creo set de datos X con n_samples (50) ejemplos en 
# Y contiene dos clases 
X, Y = make_blobs(n_samples=100, centers=2, 
                  random_state=0, cluster_std=0.50) 
```

```python
clf = svm.SVC(kernel='linear')  
```

```python
clf.fit(X,Y)
```

```python
s = clf.support_vectors_
w = clf.coef_[0]
```

```python
def plot_hyperplane(clf, min_x, max_x,offset, color):
    # get the separating hyperplane
    w = clf.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(min_x, max_x)  # me aseguro de que la linea sea lo suficientemente larga
    yy = a * xx - offset
    plt.plot(xx, yy,color, linewidth = 2)
```

```python
sv = clf.support_vectors_
```

```python
offset = (clf.intercept_[0]) / w[1]
offset1 = (-w[0] / w[1])*s[0][0]-s[0][1]
offset2 = (-w[0] / w[1])*s[2][0]-s[2][1]

# grafico scatters

fig, ax = plt.subplots(figsize=(10,10))
plt.scatter(sv[:,0],sv[:,1],s = 200, marker = 'x')
plt.scatter(X[:, 0], X[:, 1], c=Y ,s=50,alpha = 0.75, cmap='spring');
plot_hyperplane(clf , -0.5 , 3.5 , offset,'b')
plot_hyperplane(clf, -0.5, 3.5 , offset1,'m--')
plot_hyperplane(clf, -0.5, 3.5 , offset2,'y--')

plt.show()  
```
