```python
expected = y_test
predicted = clf.predict(X_test)

print("Reporte de nuestro clasificador %s:\n%s\n"
      % (clf, metrics.classification_report(expected, predicted)))
print("Matriz de confusi√≥n:\n%s" % metrics.confusion_matrix(expected, predicted))
```
