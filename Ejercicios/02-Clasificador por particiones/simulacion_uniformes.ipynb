## Enunciado

Simular $n$ puntos las distribuciones uniformes con soporte en las regiones $R_0$ y $R_1$ , donde:

- $R_0$ es el triangulo con vertices $(1,0)$, $(1,1)$ y $(\frac{1}{2},0)$
- $R_1$ es el triangulo con vertices $(0,0)$, $(\frac{1}{2},1)$ y $(0,1)$

- $n = 10, 100, 1000, 10000, \ldots$

## Solución

```python
#Import libraries
import numpy as np
#Plots
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
n = 10000 #Tamaño de muestra
```

#### Simulación de la distribucion uniforme con soporte en $R_0$

```python
muestra = np.zeros((n,3))
```

```python
y1 = np.random.uniform( 0, 1, n ) #Uniforme continua sobre (0,1)

x1 = np.zeros((n,))
for i in range(n):
    x1[i] = np.random.uniform((1-y1[i])/2 ,(1+y1[i])/2) 
```

#### Simulación de la distribucion uniforme con soporte en $R_1$

```python
i = 0 #Puntos incluidos hasta el momento.
while(i < n):
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    clase = np.random.randint( 0 , 1 + 1 ) #Uniforme discreta en {0,1}
    if (( clase == 0 and abs(2*x-1) < y) or ( clase == 1 and y < 2*x < 2-y )):
        muestra[i][0] = x
        muestra[i][1] = y
        muestra[i][2] = clase
        i+=1
```

#### Grafico las muestras

```python
clase0 , clase1 = muestra[(muestra[:,2] == 0.)] , muestra[(muestra[:,2] == 1.)]
```

```python
dimensiones = (6 , 6)
plt.figure(figsize = dimensiones)

g = plt.scatter( clase0[:,0] , clase0[:,1] , alpha='0.75', color='darkgreen' , label = 'Clase 0');
g = plt.scatter( clase1[:,0] , clase1[:,1] ,alpha='0.75', color='darkorange' , label = 'Clase 1');
plt.legend()
plt.title("Distribuciones", fontsize=20)
plt.show()
```
