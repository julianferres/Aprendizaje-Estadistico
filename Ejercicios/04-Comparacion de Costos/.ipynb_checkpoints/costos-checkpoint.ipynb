# Estimación de la función de perdida para distintos clasificadores por partición

### Julian Ferres - Nro.Padrón 101483

Sean las regiones $R_0$ y $R_1$ y la cantidad de puntos $n$, donde:

- $R_0$ es el triangulo con vertices $(1,0)$, $(1,1)$ y $(\frac{1}{2},0)$
- $R_1$ es el triangulo con vertices $(0,0)$, $(\frac{1}{2},1)$ y $(0,1)$

- $n = 10, 100, 1000, 10000, \ldots$

Se simulan $n$ puntos en $\mathbb{R}^2$ siguiendo los pasos:
    
>- Cada punto pertenece a una de las dos clases: **_Clase 0_** o **_Clase 1_** con probabilidad $\frac{1}{2}$

>- Los puntos de la clase $i$ tienen distribución uniforme con soporte en $R_i$ , con $i=0,1$

## Solución

```python
#Import libraries
import numpy as np
#Plots
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
n = 10000 #Tamaño de muestra
```

```python
train = np.zeros((n,3))
```

```python
n_test = int(n/6)
test = np.zeros((n_test,3))
```

#### Toma de muestra

```python
i = 0 #Puntos incluidos hasta el momento.
while(i < n):
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    clase = np.random.randint( 0 , 1 + 1 ) #Uniforme discreta en {0,1}
    if (( clase == 0 and abs(2*x-1) < y) or ( clase == 1 and y < 2*x < 2-y )):
        train[i][0] = x
        train[i][1] = y
        train[i][2] = clase
        i+=1
```

```python
i = 0 #Puntos incluidos hasta el momento.
while(i < n_test):
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    clase = np.random.randint( 0 , 1 + 1 ) #Uniforme discreta en {0,1}
    if (( clase == 0 and abs(2*x-1) < y) or ( clase == 1 and y < 2*x < 2-y )):
        test[i][0] = x
        test[i][1] = y
        test[i][2] = clase
        i+=1
```

```python
clase0 , clase1 = train[(train[:,2] == 0.)] , train[(train[:,2] == 1.)]
```

```python
g = plt.scatter( clase0[:,0] , clase0[:,1] , alpha='0.5', color='darkgreen' , label = 'Clase 0');
g = plt.scatter( clase1[:,0] , clase1[:,1] ,alpha='0.5', color='darkorange' , label = 'Clase 1');
plt.legend()
plt.title("Distribuciones", fontsize=20)
plt.show()
```

```python
f = lambda x : 0 if (x>= 0) else 1
f_vec = np.vectorize(f)
```

## Estimación de Perdida del set de training

### Para estimar la perdida utilizaremos el estimador de la frecuencia relativa de desaciertos

$$\widehat{L}_n( \Phi) = \frac{1}{n} \sum_{i=1}^n \mathbf{1}\{\Phi(X_i) \neq Y_i\}$$

```python
errores_1 = 0
for i in range(n):
    
    y_p = int(train[i,1]>=1/2)
    errores_1 = errores_1 + 1 if particion1[y_p , 0] != train[i,2] else errores_1
```

```python
errores_1/n
```

```python
errores_2 = 0
for i in range(n):
    
    x_p = int(train[i,0]>=1/2)
    
    errores_2 = errores_2 + 1 if particion2[0,x_p] != train[i,2] else errores_2
```

```python
errores_2/n
```

```python
errores_3 = 0

for i in range(n):
    
    casilla = int(train[i,0]>train[i,1])
    
    errores_3 = errores_3 + 1 if particion3[casilla] != train[i,2] else errores_3
```

```python
errores_3/n
```

```python
errores_4 = 0
for i in range(n):
    
    casilla = int(1-train[i,0]<train[i,1])
    
    errores_4 = errores_4 +1 if particion4[casilla] != train[i,2] else errores_4
```

```python
errores_4/n
```

```python
errores_5 = 0
for i in range(n):
    
    casilla = int(1-train[i,0]<train[i,1]) + 2*int(train[i,0]>train[i,1])
    
    errores_5 = errores_5 +1 if particion5[casilla] != train[i,2] else errores_5
```

```python
errores_5/n
```

```python
errores_6 = 0
for i in range(n):
    
    x_p, y_p = int(train[i,0]>=1/2) , int(train[i,1]>=1/2)
    
    errores_6 = errores_6+1 if particion6[y_p , x_p] != train[i,2] else errores_6
```

```python
errores_6/n
```

## Estimación de Perdida del set de testing

### Para estimar la perdida utilizaremos el estimador de la frecuencia relativa de desaciertos

$$\widehat{L}_n( \Phi) = \frac{1}{n_{test}} \sum_{i=1}^{n_{test}} \mathbf{1}\{\Phi(X_i) \neq Y_i\}$$

```python
err_1_test = 0
for i in range(n_test):
    
    y_p = int(test[i,1]>=1/2)
    err_1_test = err_1_test + 1 if particion1[y_p , 0] != test[i,2] else errores_1_test
```

```python
err_1_test/n_test
```

```python
err_2_test = 0
for i in range(n_test):
    
    x_p = int(test[i,0]>=1/2)
    
    err_2_test = err_2_test + 1 if particion2[0, x_p] != test[i,2] else err_2_test
```

```python
err_2_test/n_test
```

```python
err_3_test = 0

for i in range(n_test):
    
    casilla = int(test[i,0]>test[i,1])
    
    err_3_test = err_3_test + 1 if particion3[casilla] != train[i,2] else err_3_test
```

```python
err_3_test/n_test
```

```python
err_4_test = 0
for i in range(n_test):
    
    casilla = int(1-test[i,0]<test[i,1])
    
    err_4_test = err_4_test +1 if particion4[casilla] != train[i,2] else err_4_test
```

```python
err_4_test/n_test
```

```python
err_5_test = 0
for i in range(n_test):
    
    casilla = int(1-test[i,0]<test[i,1]) + 2*int(test[i,0]>test[i,1])
    
    err_5_test = err_5_test +1 if particion5[casilla] != train[i,2] else err_5_test
```

```python
err_5_test/n_test
```

```python
err_6_test = 0
for i in range(n_test):
    
    x_p, y_p = int(test[i,0]>=1/2) , int(test[i,1]>=1/2)
    
    err_6_test = err_6_test+1 if particion6[y_p , x_p] != train[i,2] else err_6_test
```

```python
err_6_test/n_test
```
